## Python Variables

#### A variable is a name used to store a value in memory. In Python, you do not need to declare the data type explicitly — the type is automatically determined based on the assigned value.

```python
print("Hello World")
```

### Example 1: Basic Variable

```python
student_name = "Abhishek"
year = 3
cgpa = 8.2

print(student_name)
print(year)
print(cgpa)
```

### Example 2: Type Checking


```python
x = 100
y = 9.5
z = "Python"

print(type(x))
print(type(y))
print(type(z))
```

### Example 3: Dynamic Typing

```python
data = 50
print(data)
data = "fifty"
print(data)
```

#### The same variable can hold different types of values at different times.

### Example 4: Multiple Assignment

```python
a, b, c = 10, 20, 30
print(a, b, c)
```

### Example 5: Assigning Same Value

```python
p = q = r = 1
print(p, q, r)
```

### Example 6: Swapping Values

```python
x = 5
y = 10
x, y = y, x
print("x:", x, "y:", y)
```

### Example 7: Updating a Variable

```python
count = 5
count = count + 3
print(count)
```

### Example 8: Real-Life Example

```python
price = 250
quantity = 2
total_amount = price * quantity
print("Total:", total_amount)
```

### Example 9: User Input

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))

print(f"{name} is {age} years old")
```

### Example 10: Naming Rules

```python
user_name = "ok"   # valid
age1 = 20          # valid

# invalid examples:

#           1name = "wrong"
#         user-name = "wrong"
#           class = "wrong"
```

## Key Points
- A variable stores a value  
- No need to declare data type explicitly  
- The type can change during execution  
- Use meaningful and readable variable names.
